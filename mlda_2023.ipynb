```python
# Кросс-валидация
from sklearn.model_selection import KFold
from tqdm import tqdm 
import catboost as cb
from sklearn.metrics import roc_auc_score


oof_fc = submission.copy()

kf = KFold(n_splits=4, shuffle=True, random_state=42)

feature_lst = num_features + cat_features + add_features + text_features

for i, (train_index, test_index) in enumerate(kf.split(train)):
    
    print('_'*100)
    print()
    print(f'Фолд {i} запущен')
    
    X_train = train.loc[train_index, feature_lst]
    X_val = train.loc[test_index, feature_lst]
    y_train = train.loc[train_index, target]
    y_val = train.loc[test_index, target]

    
    model = cb.CatBoostClassifier(
        cat_features=cat_features, 
        text_features=text_features,
        verbose=False,
        reg_lambda=5,
        ignored_features=ignore
    )
    model.fit(X_train, y_train)
    
    oof_fc[f'prediction_{i}'] = model.predict_proba(test[feature_lst])[:, 1]
    
    auc = roc_auc_score(y_val, model.predict_proba(X_val[feature_lst])[:, 1])
    print(f'AUC: {np.round(auc, 3)}')
    
    print()
    print('_'*100)
    
```
